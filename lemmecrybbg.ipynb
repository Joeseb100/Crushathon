```python
# --- Imports ---
import torch
import re # Import regular expressions
from transformers import AutoTokenizer, AutoModelForCausalLM # Or your specific model class

# --- Prerequisites ---
device = torch.device("cuda" if torch.cuda.is_available() else "cpu")
print(f"Using device: {device}")

# --- CONFIGURATION ---
# <<<--- Replace with your actual model name --->>>
model_name = "gpt2" # Example: Use "microsoft/DialoGPT-medium" for a conversational model
# <<<------------------------------------------->>>

# Load tokenizer and model
print(f"Loading model and tokenizer: {model_name}")
tokenizer = AutoTokenizer.from_pretrained(model_name)
model = AutoModelForCausalLM.from_pretrained(model_name)

# --- Set Pad Token (Crucial) ---
# Set pad_token if needed (often needed for models like GPT-2 if using padding/attention mask)
if tokenizer.pad_token is None:
    if tokenizer.eos_token is not None:
        print(f"Tokenizer lacks pad_token. Setting pad_token to eos_token ({tokenizer.eos_token}).")
        tokenizer.pad_token = tokenizer.eos_token
    else:
        # If EOS is also missing, add a generic pad token (less common)
        print("Tokenizer lacks both pad_token and eos_token. Adding a '[PAD]' token.")
        tokenizer.add_special_tokens({'pad_token': '[PAD]'})
        # IMPORTANT: If you add a *new* token, you MUST resize model embeddings
        model.resize_token_embeddings(len(tokenizer))
        print("Resized model token embeddings.")

# Move model to device and set to evaluation mode
model.to(device)
model.eval()
print("Model and tokenizer loaded successfully.")
# --- End of Prerequisites ---


# --- Chat Function (Handles generation with history and attention mask) ---
def chat_with_bot(user_input, chat_history_ids=None, chat_history_mask=None):
    """
    Handles one turn of the conversation, including attention mask.
    """
    # 1. Encode new user input, getting both IDs and attention mask
    # Add EOS token to signal end of user turn
    encoded_input = tokenizer(user_input + tokenizer.eos_token, return_tensors='pt')
    new_input_ids = encoded_input['input_ids'].to(device)
    new_attention_mask = encoded_input['attention_mask'].to(device)

    # 2. Prepend the chat history and attention mask (if any)
    if chat_history_ids is not None:
        # Ensure history is on the correct device
        chat_history_ids = chat_history_ids.to(device)
        chat_history_mask = chat_history_mask.to(device)
        # Concatenate history with new input
        bot_input_ids = torch.cat([chat_history_ids, new_input_ids], dim=-1)
        attention_mask = torch.cat([chat_history_mask, new_attention_mask], dim=-1)
    else:
        # First turn
        bot_input_ids = new_input_ids
        attention_mask = new_attention_mask

    # --- Context Truncation (Handles IDs and Mask) ---
    # Use model's config for max length, provide a default if not found
    max_model_len = getattr(model.config, 'max_position_embeddings', 1024)
    current_len = bot_input_ids.shape[-1]
    # Define how many tokens to reserve for the response generation
    generation_reserve = 150 # Allow up to 150 tokens for the bot's reply

    if current_len > max_model_len:
        print(f"Warning: Combined input length ({current_len}) exceeds model max length ({max_model_len}). Truncating...")
        # Calculate how many tokens to keep from the history + new input
        keep_length = max_model_len - generation_reserve
        if keep_length <= 0: # Safety check if reserve is too large
             keep_length = max_model_len // 2 # Fallback: keep half
        # Truncate from the left (oldest history)
        bot_input_ids = bot_input_ids[:, -keep_length:]
        attention_mask = attention_mask[:, -keep_length:] # TRUNCATE MASK TOO!
        print(f"Truncated input length to: {bot_input_ids.shape[-1]}")
        current_len = bot_input_ids.shape[-1] # Update current_len after truncation
    # --- End of Context Truncation ---

    # 3. Generate a response using the model
    print(f"Generating response from input shape: {bot_input_ids.shape}")
    with torch.no_grad():
        outputs = model.generate(
            input_ids=bot_input_ids,
            attention_mask=attention_mask, # Pass the combined attention mask
            max_length=min(current_len + generation_reserve, max_model_len), # Limit total length
            pad_token_id=tokenizer.pad_token_id, # Use the designated pad token ID
            temperature=0.7,   # Control randomness
            top_k=50,          # Consider top K tokens
            top_p=0.95,        # Use nucleus sampling
            do_sample=True,    # Enable sampling
            eos_token_id=tokenizer.eos_token_id # Help model stop generating
        )
    print(f"Output history shape: {outputs.shape}")

    # The generated 'outputs' tensor is the new full chat history (IDs)
    new_chat_history_ids = outputs

    # Create a new attention mask corresponding to the *entire* output history.
    # Since generate typically doesn't add padding itself when a mask is provided,
    # we can assume all tokens in the output are real and should be attended to.
    new_chat_history_mask = torch.ones_like(new_chat_history_ids).to(device)

    # 4. Decode only the newly generated part of the response
    # Find the length of the original input to slice the output correctly
    original_input_length = bot_input_ids.shape[-1]
    response_ids = new_chat_history_ids[:, original_input_length:]
    response = tokenizer.decode(response_ids[0], skip_special_tokens=True)

    # Return the response string and the updated history tensors
    return response, new_chat_history_ids, new_chat_history_mask


# --- Safety Check Configuration ---
# <<<--- Expand this list significantly for better coverage! --->>>
crisis_keywords = [
    "suicid", "kill myself", "want to die", "self harm", "self-harm",
    "hopeless", "end my life", "cut myself", "overdose", "no reason to live"
    ]
# <<<------------------------------------------------------------>>>
print(f"Initialized with {len(crisis_keywords)} crisis keywords for safety check.")


# --- Main Interaction Loop ---
chat_history_ids = None   # Initialize history IDs
chat_history_mask = None  # Initialize history mask
print("\nTherapyBot initialized. Type 'quit' to exit.")
print("Disclaimer: This bot is NOT a substitute for professional mental health support.")

while True:
    user = input("You: ")
    if user.lower() == "quit":
        print("TherapyBot: Goodbye! Take care.")
        break

    # --- Safety Check Integration ---
    user_lower = user.lower()
    # Use word boundaries (\b) in regex for more precise matching (optional but recommended)
    # Example: \bsuicid\b would match "suicide" but not "suicidal ideation" if you only want the former
    # Simple substring check for now:
    if any(keyword in user_lower for keyword in crisis_keywords):
        print("\nTherapyBot: I hear that you're going through immense pain right now, and it takes courage to talk about it. "
              "It's really important to reach out for professional help when feeling this way. "
              "You can connect with people who can support you by calling or texting 988 anytime in the US and Canada. In the UK, you can call 111. "
              "These services are free, confidential, and available 24/7. Please reach out to them.")
        # Optional: Reset history if you want the bot to forget the context after a crisis message
        # print("(Resetting conversation history for safety.)")
        # chat_history_ids = None
        # chat_history_mask = None
        continue # Skip the model generation for this turn and ask for new input
    else:
        # --- Proceed with calling chat_with_bot if no crisis keywords found ---
        try:
            # Get reply and update history (both IDs and mask)
            reply, chat_history_ids, chat_history_mask = chat_with_bot(
                user,
                chat_history_ids,
                chat_history_mask
            )
            print("TherapyBot:", reply.strip()) # Use strip() to remove leading/trailing whitespace

        except Exception as e:
            print(f"\nAn error occurred during generation: {e}")
            print("There was an issue generating a response. Resetting conversation history.")
            # Reset history on error to prevent corrupted state
            chat_history_ids = None
            chat_history_mask = None
            # Consider logging the error in a real application: import logging; logging.exception("Error during chat")
```

```python
# --- Ensure these imports are at the TOP of your script/notebook cell ---
import torch
from transformers import AutoTokenizer, AutoModelForCausalLM # Or your specific model class

# --- Prerequisites (Ensure these are executed first) ---
device = torch.device("cuda" if torch.cuda.is_available() else "cpu")
print(f"Using device: {device}")

# Replace with your actual model name
model_name = "microsoft/DialoGPT-medium" # e.g., "gpt2", "DialoGPT-medium"

# Load tokenizer and model
tokenizer = AutoTokenizer.from_pretrained(model_name)
model = AutoModelForCausalLM.from_pretrained(model_name) # Or your specific model class

# Set pad_token if needed (Crucial for the warning!)
if tokenizer.pad_token is None:
    print("Setting pad_token = eos_token")
    tokenizer.pad_token = tokenizer.eos_token
    # Note: If you added a *new* pad token instead, you'd need to resize embeddings
    # tokenizer.add_special_tokens({'pad_token': '[PAD]'})
    # model.resize_token_embeddings(len(tokenizer))

# Move model to device and set to evaluation mode
model.to(device)
model.eval()
print("Model and tokenizer loaded.")
# --- End of Prerequisites ---


# --- Updated Chat Interaction Code ---

# Keep track of both IDs and Attention Mask history
def chat_with_bot(user_input, chat_history_ids=None, chat_history_mask=None):
    """
    Handles one turn of the conversation, now including attention mask.

    Args:
        user_input (str): The text input from the user.
        chat_history_ids (torch.Tensor, optional): Tensor of token IDs for history. Defaults to None.
        chat_history_mask (torch.Tensor, optional): Tensor of attention masks for history. Defaults to None.

    Returns:
        tuple: (str, torch.Tensor, torch.Tensor)
            - The bot's response string.
            - The updated chat history token IDs tensor.
            - The updated chat history attention mask tensor.
    """
    # 1. Encode new user input. The tokenizer directly gives us the attention mask here!
    # We no longer manually add eos_token here, let generate handle context/stops.
    # Encoding *only* the user input for this turn.
    encoded_input = tokenizer(user_input + tokenizer.eos_token, return_tensors='pt')
    new_input_ids = encoded_input['input_ids'].to(device)
    new_attention_mask = encoded_input['attention_mask'].to(device)

    # 2. Prepend the chat history and attention mask (if any)
    if chat_history_ids is not None:
        # Move history to the correct device if needed
        chat_history_ids = chat_history_ids.to(device)
        chat_history_mask = chat_history_mask.to(device)
        # Concatenate history with new input
        bot_input_ids = torch.cat([chat_history_ids, new_input_ids], dim=-1)
        attention_mask = torch.cat([chat_history_mask, new_attention_mask], dim=-1)
    else:
        # First turn, no history yet
        bot_input_ids = new_input_ids
        attention_mask = new_attention_mask

    # --- Optional: Context Truncation (Now includes mask) ---
    max_model_len = getattr(model.config, 'max_position_embeddings', 1024)
    current_len = bot_input_ids.shape[-1]

    if current_len > max_model_len:
        print(f"Warning: Input length ({current_len}) exceeds model max length ({max_model_len}). Truncating...")
        keep_length = max_model_len - 150 # Keep space for generation
        bot_input_ids = bot_input_ids[:, -keep_length:]
        attention_mask = attention_mask[:, -keep_length:] # TRUNCATE MASK TOO!
        print(f"Truncated input length to: {bot_input_ids.shape[-1]}")
    # --- End of Context Truncation ---

    # 3. Generate a response using the model, explicitly passing the attention mask
    print(f"Generating response from input shape: {bot_input_ids.shape}")
    with torch.no_grad():
        # Pass both input_ids and attention_mask
        outputs = model.generate(
            input_ids=bot_input_ids,
            attention_mask=attention_mask, # <-- THE FIX FOR THE WARNING
            max_length=min(bot_input_ids.shape[-1] + 150, max_model_len), # Limit total len dynamically
            pad_token_id=tokenizer.pad_token_id, # Use the designated pad token ID
            temperature=0.7,
            top_k=50,
            top_p=0.95,
            do_sample=True,
            eos_token_id=tokenizer.eos_token_id # Helps model know when to stop
        )
    print(f"Output history shape: {outputs.shape}")

    # The generated output 'outputs' tensor now IS the new chat_history_ids
    # We need to generate the corresponding attention mask for this new history
    # Since generate doesn't pad internally *in the same way* when we provide a mask,
    # the output length should match the actual tokens. We can assume all are attended.
    new_chat_history_ids = outputs
    # Create a matching attention mask (all 1s up to the length of the output)
    new_chat_history_mask = torch.ones_like(new_chat_history_ids).to(device)


    # 4. Decode the generated response (only the new part)
    response_ids = new_chat_history_ids[:, bot_input_ids.shape[-1]:]
    response = tokenizer.decode(response_ids[0], skip_special_tokens=True)

    # Return the response and the *full* updated history tensors
    return response, new_chat_history_ids, new_chat_history_mask


# --- Main Interaction Loop (Updated) ---
chat_history_ids = None   # Initialize history IDs
chat_history_mask = None  # Initialize history mask
print("TherapyBot initialized. Type 'quit' to exit.")

while True:
    user = input("You: ")
    if user.lower() == "quit":
        print("TherapyBot: Goodbye!")
        break

    try:
        # Get reply and update history (both IDs and mask)
        reply, chat_history_ids, chat_history_mask = chat_with_bot(
            user,
            chat_history_ids,
            chat_history_mask
        )
        print("TherapyBot:", reply)

    except Exception as e:
        print(f"An error occurred: {e}")
        # Consider more specific error handling if needed
        print("Resetting conversation history due to error.")
        chat_history_ids = None # Reset history on error
        chat_history_mask = None
```
